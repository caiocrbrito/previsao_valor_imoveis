# Análise de Componentes Principais (PCA)

```python
from sklearn.decomposition import PCA
```

```python
def func_norm(x):
    return (x - x.mean()) / x.std()
```

```python
# Criando uma copia dos dados
df_pca = df[['quartos',
            'suites',
            'vagas',
            'espaco',
            'latitude',
            'longitude']]
```

```python
# Padronizando os dados continuos
df_pca['espaco'] = func_norm(df_pca['espaco'])
df_pca['latitude'] = func_norm(df_pca['latitude'])
df_pca['longitude'] = func_norm(df_pca['longitude'])

# y_train = (y_train - y_train.mean()) / y_train.std()
```

```python
# Realizar a análise de PCA
pca = PCA()
pca.fit(df_pca)# Obter as proporções da variância explicada
explained_variance_ratio = pca.explained_variance_ratio_
```

```python
# Plotar a variância explicada acumulada
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)
plt.plot(range(1, len(cumulative_variance_ratio) + 1), cumulative_variance_ratio)
plt.xlabel('Número de Componentes Principais')
plt.ylabel('Variância Explicada Acumulada')
plt.title('Análise de Componentes Principais (PCA)')
plt.show();

# Obter os coeficientes do PCA (cargas)
loadings = pca.components_

# Mostrar os coeficientes e suas respectivas variáveis
loadings_df = pd.DataFrame(loadings.T, columns=['PC{}'.format(i+1) for i in range(len(loadings))], index=df_pca.columns)
print(loadings_df)
```

A analise por componentes principais mostra que com 4 variaveis temos mais de 95% da explicação do modelo e que a latitude e longitude, que nos testes de correlação se mostraram insignificantes, conjuntamente, estão entre as 3 variáveis mais importantes para previsão do modelo. Como as variáveis são altamente correlacionadas, devemos escolher com cuidado as mais importantes e não usar todas, para que não haja problemas como overfitting.

Também é possível visualizar a relação entre localidade e o valor dos imóveis graficamente.

```python
from sklearn.cluster import KMeans
import geopandas as gpd

# Selecionar as variáveis relevantes
variables = df[['valor_m2', 'latitude', 'longitude']]

# Normalizar as variáveis
normalized_df = (variables - variables.mean()) / variables.std()

# Definir o número de clusters
num_clusters = 10

# Executar o algoritmo de agrupamento (K-means)
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(normalized_df)

# Obter os rótulos de cluster para cada amostra
cluster_labels = kmeans.labels_

# Adicionar os rótulos de cluster ao DataFrame original
df['cluster'] = cluster_labels

# Converta o DataFrame para um GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['longitude'], df['latitude']))

# Plotar o mapa
plt.figure(figsize=(10, 10))
ax = gdf.plot(column='cluster', cmap='viridis', legend=True)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Agrupamento de Valor_m2, Latitude e Longitude')

plt.show();
```

# Regressão linear multipla

## Regressão linear com XGBoost

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import ensemble
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.model_selection import cross_val_score, KFold
```

```python
df_dummies = pd.get_dummies(df[['preco', 'suites', 'espaco', 'latitude', 'longitude']], columns=['suites']).drop(columns='suites_5')
```

```python
X = df_dummies.iloc[:,1:].values
y = df_dummies.iloc[:,0].values
```

```python
# Separando o dataframe em dados de treino e dados de teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Padronizando os dados continuos
scaler = StandardScaler()

X_train[:, 0:3] = scaler.fit_transform(X_train[:, 0:3])
X_test[:, 0:3] = scaler.transform(X_test[:, 0:3])

y_train = scaler.fit_transform(y_train.reshape(-1,1))
y_test = scaler.transform(y_test.reshape(-1,1))
```

```python
# Treinando o modelo com XGBoost
reg = ensemble.GradientBoostingRegressor()
model = reg.fit(X_train, y_train)
```

```python
# Imprimindo o score de treinamento
model.score(X_train, y_train)
```

```python
# Imprimindo o score de teste
model.score(X_test, y_test)
```

```python
# Fazendo cross validation para avaliar o modelo em diversas amostras
kfold = KFold(n_splits=15, shuffle=True, random_state=5)
resultado = cross_val_score(model, X_test, y_test, cv=kfold)
print('Acurácia Média: %.2f%%' % (resultado.mean() * 100.0))
```

Analizando o modelo através da técnica de validação cruzada, a acurácia do modelo foi de 89.83%, o que é um valor muito bom. Mas esta acurácia sozinha não é garantia de um bom modelo, pois a regressão linear deve seguir quatro pressupostos, a linearidade entre as variáveis independentes e dependente, homocedasticidade, independência e normalidade dos resíduos. 

## Regressão Linear com SkLearn

A regressão linear é uma técnica muito simples, quando bem compreendida. Ao se aplicar um algoritmo de regressão, o resultado é uma equação que calcula a melhor reta que se adapta ao conjunto de pontos e adapta novos inputs em algum lugar desta reta. O cálculo da reta que melhor se adapta aos pontos não é complicado, resultando em valores próximos seja qual for o algoritmo que utilizar. Como não tenho acesso aos valores calculados com XGBoost, criarei uma regressão linear simples para comparar com os valores fornecidos com statsmodels.

```python
from sklearn import linear_model
```

```python
reg2 = linear_model.LinearRegression()
model2 = reg2.fit(X_train, y_train)
```

```python
# Coeficientes do modelo
print('Valor de alpha:', reg2.intercept_[0])
print('Valores dos betas:', reg2.coef_[0])
```

```python
kfold = KFold(n_splits=15, shuffle=True, random_state=5)
resultado2 = cross_val_score(model2, X_test, y_test, cv=kfold)
print('Acurácia Média: %.2f%%' % (resultado2.mean() * 100.0))
```

O valor de acurácia ficou bem próximo da acurácia do XGBoost. Provavelmente possuem valores de betas muito próximos, como mencionado anteriormente.

## Criando e avaliando uma regressão com statsmodels

```python
import statsmodels.api as sm
import statsmodels
import statsmodels.formula.api as smf
```

```python
df_stats = df_dummies.copy()
```

```python
# Normalizando os dados continuos
df_stats.preco = func_norm(df_stats.preco)
df_stats.espaco = func_norm(df_stats.espaco)
df_stats.latitude = func_norm(df_stats.latitude)
df_stats.longitude = func_norm(df_stats.longitude)
```

```python
df_stats
```

```python
# Criando o modelo de regressao
res = smf.ols(formula='preco ~ \
                        espaco + \
                        latitude + \
                        longitude + \
                        suites_0 + \
                        suites_1 + \
                        suites_2 + \
                        suites_3 + \
                        suites_4', 
              data=df_stats).fit()
```

```python
res.summary()
```

```python
# Plot dos resíduos do modelo
plt.scatter(res.predict(), res.resid)
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Residual plot')
plt.show()
```

Visualmente, é possível ver uma tendência em forma de cone. A dispersão dos dados sugere que o modelo preve muito bem valores no primeiro desvio padrão para ambos os lados, mas para valores acima do primeiro desvio padrão, a previsão não será muito boa. Isso se deve à distribuição dos dados, como o esperado, há uma maior quantidade de imóveis em uma faixa mais baixa à intermediária de preços. É possível resolver este problema através de uma transformação linear ou aplicação de alguma técnica de balanceamento ou simplesmente excluindo dados outliers. Dependendo do objetivo final, devemos tomar a decisão mais apropriada.

```python
# Execute o teste de Breusch-Pagan para heterocedasticidade
from statsmodels.stats.diagnostic import het_breuschpagan

bp = het_breuschpagan(res.resid, res.model.exog)

# Imprimindo o valor do teste
print(bp)
```

O teste de Breusch-Pagan mostra um p-valor muito pequeno, o que indica uma rejeição da hipotese nula e há heterocedasticidade, que é uma correlação muito forte entre variáveis preditivas, que influenciam negativamente a qualidade do modelo. Deve ser escolhida alguma das variáveis preditoras para ser eliminada do modelo.

```python
# Calculo do VIF (Variance Inflation Factors) para cada variável independente
from statsmodels.stats.outliers_influence import variance_inflation_factor

vifs = [variance_inflation_factor(res.model.exog, i) for i in range(len(res.params))]

# Imprimindo os valores do VIF
print(vifs)
```

Os valores do teste VIF mostram que o intercepto e os valores de quantidade de suites estão altamente correlacionados entre si. Isso pode levar o modelo a overfitting e vieses, não retratando a realidade esperada. Alguma variável deve ser retirada do modelo, o ideal é retirar espaço mantendo suites e retirar suites mantendo espaço e avaliar novamente como o modelo se comporta. 

```python
# qqplot
from statsmodels.graphics.gofplots import qqplot

qqplot(res.resid, line='45')
plt.show()
```

O qqplot mostra a normalidade dos resíduos. Quando os resíduos apresentam uma distribuição aproximadamente normal, os pontos devem estar sobre a reta ou muito próximos disto. Como os resíduos não apresentam uma distribuição normal, as estatísticas de saída do modelo não são válidas e não é possível afirmar se o modelo e suas variáveis são relevantes, já que as estatísticas assumem a normalidade dos resíduos para serem calculadas.
