# Scraping das informações

```python
# importando as bibliotecas
import pandas as pd
import requests
from bs4 import BeautifulSoup
from time import sleep
```

```python
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'}
```

```python
lista = []
links = []

for n in range(1, 168):

    url = f'https://www.dfimoveis.com.br/venda/df/aguas-claras/apartamento?pagina={n}'
    content = requests.get(url, headers=header)
    
    sleep(1.5)

    soup = BeautifulSoup(content.content, 'html.parser')

    for link in soup.find('div', {'id': 'resultadoDaBuscaDeImoveis'}).findAll('a', href=True):
        
        try:
            titulo = link.find('h2', {'class': 'new-title phrase'}).getText()
            desc = link.find('h3', {'class': 'new-desc phrase'}).getText()
            desc2 = link.find('h3', {'class': 'new-simple phrase'}).getText()
            preco = link.find('div', {'class': 'new-price'}).findAll('h4')[0].getText()
            valor_m = link.find('div', {'class': 'new-price'}).findAll('h4')[1].getText()
            esp = link.find('ul', {'class': 'new-details-ul'}).getText()
            desc_local = link.find('div', {'class': 'new-text phrase'}).getText()
            link_anun = link.get('href')

            lista.append([titulo, desc, desc2, preco, valor_m, esp, desc_local, link_anun])
        except:
            links.append((link.get('href')))
```

```python
print(len(lista), 'ok')
print(len(links), 'fail')
```

```python
df = pd.DataFrame(lista, columns=['titulo', 'desc', 'desc2', 'preco', 'valor_m2', 'espec', 'desc3', 'link_anuncio'])
df
```

```python
df['conteudo'] = ''
```

```python
for i in range(len(df.link_anuncio)):
    content = requests.get('https://www.dfimoveis.com.br'+df.link_anuncio[i])
    soup = BeautifulSoup(content.content, 'html.parser')

    df.loc[i, 'conteudo'] = ' '.join([j.getText() for j in soup.findAll('div', {'class': 'pr-md-2 aumentar-tela'})])
```

```python
df.to_csv('/content/drive/MyDrive/Trabalhos/imoveis/imoveis.csv', sep=';', index=False, encoding='utf-8')
```
